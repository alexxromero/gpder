# Gaussian Process Regression

This demo shows how to use the GaussianProcessRegressor class to regress the the dynamical system 

$y = \sin(3x) + \cos(x)^2 + \epsilon$  
$\frac{dy}{dx} = 3\cos(2x) - 2\cos(x)\sin(x) + \epsilon$

Where the function and derivative observations have white noise with variance $\epsilon ~ \mathcal{N}(0, 0.2)$.

```python
import sys 
import numpy as np
from IPython.display import IFrame
import matplotlib.pyplot as plt
%matplotlib inline  

from sklearn.metrics import mean_squared_error

sys.path.append("../../../gpder")
import gpder
from gpder.gaussian_process import GaussianProcessRegressor
from gpder.gaussian_process.kernels import GPKernel, GPKernelDerAware

from plotting_utils import plot_gp
```

```python
def F(X, e=0.2):
    noise = np.random.normal(0, e, X.shape)
    return X * np.sin(2 * X) + noise

def dF(X, e=0.2):
    noise = np.random.normal(0, e, X.shape)
    return np.sin(2 * X) + 2 * X * np.cos(2 * X) + noise
```

```python
np.random.seed(123)
X_lower, X_upper = -2*np.pi, 2*np.pi

# -- testing points -- #
X_test = np.linspace(X_lower, X_upper, 100).reshape(-1, 1)
y_test = F(X_test, e=0)
dX_test = X_test
dy_test = dF(dX_test, e=0)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].plot(X_test, y_test, color='darkviolet', 
           label='noise-free \n function observations')
ax[0].set_xlabel(r'$x$', fontsize=18)
ax[0].set_ylabel(r'$y$', fontsize=18)
ax[0].set_title('Function Observations', fontsize=18)
ax[0].tick_params(axis='both', which='major', labelsize=14)
ax[1].plot(dX_test, dy_test, color='darkgreen', 
           label='noise-free \n derivative observations')
ax[1].set_xlabel(r'$x$', fontsize=18)
ax[1].set_ylabel(r'${dy}/{dx}$', fontsize=18)
ax[1].set_title('Derivative Observations', fontsize=18)
ax[1].tick_params(axis='both', which='major', labelsize=14)
# lines_labels = [ax.get_legend_handles_labels() for ax in fig.axes]
# lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]
# fig.legend(lines, labels, fontsize=16, ncol=2,
#            loc=(0.2, 0.9))
plt.show()
```

## Regular GPR

We perform five experiments to observe the function observarions. These experiments are used to train the GP regressor. 

```python
# -- training points -- #
#X_train = np.random.uniform(X_lower, X_upper, 5).reshape(-1, 1)
X_train = np.linspace(X_lower, X_upper, 5).reshape(-1, 1)
y_train = F(X_train)
```

```python
kernel = GPKernel(constant_value=1, length_scale=0.5, noise_level=0.02)
gp = GaussianProcessRegressor(kernel=kernel, optimizer=None)
gp.fit(X_train, y_train)
print("Kernel: ", gp.kernel_)

y_pred, cov = gp.predict(X_test.reshape(-1, 1), return_cov=True)
_, std = gp.predict(X_test.reshape(-1, 1), return_std=True)

mse = mean_squared_error(y_test, y_pred)
print("MSE : ", mse)
print("Uncert : ", np.trace(cov))

plot_gp(gp, X_test, nsamples=3)
```

## Derivative GPR

We now include derivative observations.

```python
# -- training points -- #
dX_train = X_train
dy_train = dF(dX_train)
```

```python
kernel = GPKernelDerAware(constant_value=1, length_scale=0.5,
                          noise_level=0.02, noise_level_dX=0.02)
gp = GaussianProcessRegressor(kernel=kernel, optimizer=None)
gp.fit(X_train, y_train, dX_train, dy_train)
print("Kernel: ", gp.kernel_)

y_pred, cov = gp.predict(X_test, return_cov=True)
_, std = gp.predict(X_test, return_std=True)
mse = mean_squared_error(y_test, y_pred)
print(" ----- Function obs ----- ")
print("MSE : ", mse)
print("Uncert : ", np.trace(cov))
print(" ")

dy_pred, dy_cov = gp.predict_der(X_test.reshape(-1, 1), return_cov=True)
_, dy_std = gp.predict_der(X_test.reshape(-1, 1), return_std=True)
dy_mse = mean_squared_error(dy_test, dy_pred)
print(" ----- Derivative obs ----- ")
print("MSE : ", dy_mse)
print("Uncert : ", np.trace(dy_cov))

plot_gp(gp, X_test, nsamples=3)
```
